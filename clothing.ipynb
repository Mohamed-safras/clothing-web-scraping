```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import json
import csv
import os
import numpy as np
```

```python
BASE_URL = "https://www.flipkart.com"
listof_cloths = []
HEADERS = ({'User-Agent':'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/44.0.2403.157 Safari/537.36','Accept-Language': 'en-US'})

def get_data(catagory, page):
    
 
    relative_path = "{}/{}&page{}".format(BASE_URL, catagory, page)
    response = requests.get(relative_path,HEADERS)
    soup = BeautifulSoup(response.content, "html.parser")
    
    results = soup.find_all("div",class_="_1xHGtK _373qXS")
    for result in results:
        
        try:
            product_image = result.find("img",class_="_2r_T1I").get("src")
        except:
            product_image = None
        try :
            brand = result.find("div",class_="_2WkVRV").get_text()
        except:
            brand = None
        try:
            product_title = result.find("a",class_="IRpwTa").get_text()
        except :
            product_title = None
        try:
            new_price = result.find("div",class_="_30jeq3").get_text()
        except:
            new_price = None
        try:
            old_price = result.find("div",class_="_3I9_wc").get_text()
        except:
            old_price = None
            
        cloth = {"title":product_title,"brand":brand,"image":product_image,"new_price":new_price,"old_price":old_price}
        
        listof_cloths.append(cloth)
    
```

```python
for page in range(100):
    get_data("clothing-and-accessories/topwear/tshirt/men-tshirt/pr?sid=clo%2Cash%2Cank%2Cedy&otracker=categorytree&otracker=nmenu_sub_Men_0_T-Shirts", page)      
```
